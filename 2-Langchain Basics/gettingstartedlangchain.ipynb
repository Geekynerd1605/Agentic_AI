AI Message on Invoking a model:
Any output givem by an LLM is called AI message in form of class. Answers given by LLMs will be called as AI message.

PROMPT ENGINEERING:
Providing an instruction to LLM on how it should behave.

```python
# PROMPT ENGINEERING:
from langchain_core.prompts import ChatPromptTemplate 
# prompt template is like a class where we can define our own prompts and ask llms to behave like that

prompt=ChatPromptTemplate.from_messages(
    [
        ('system', "You are an expert AI Engineer, provide me answer based on the question."),
        ('user', '{input}')
    ]
)
prompt
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model='llama-3.1-8b-instant')
model
```

```python
### Chaining -> chaining combines the prompt to the LLM. 
### Order followed while chaining -> prompt, model, output
chain=prompt|model
chain
```

```python
response=chain.invoke({'input':'Can you tell me something about langsmith?'})
print(response)
```

```python
print(response.content)
```

```python
# Inside chaining, we can also define Output Parser (way to display output)

# OUTPUT PARSER
from langchain_core.output_parsers import StrOutputParser

output_parser=StrOutputParser()
chain = prompt|llm|output_parser
response=chain.invoke({'input': 'Hi, can you tell me about Langsmith?'})
print(response)
```
