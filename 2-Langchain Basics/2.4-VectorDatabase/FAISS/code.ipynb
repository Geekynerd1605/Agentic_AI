```python
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
embeddings
```

```python
embeddings.embed_query("Hello AI")
# O/P is vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
documents=["what is the capital of USA?",
           "Who is the president of USA?",
           "Who is the Prime Minister of India?"]
```

```python
my_query="Narendra Modi is the Prime Minister of India."
```

```python
document_embedding=embeddings.embed_documents(documents)
document_embedding
```

```python
query_embedding=embeddings.embed_query(my_query)
query_embedding
```

```python
## COSINE SIMILARITY CHECK

cosine_similarity([query_embedding], document_embedding)
```

```python
### Euclidean Distance ==> checks the distances b/w 2 vectors --> lowest distance is more closest
from sklearn.metrics.pairwise import euclidean_distances
```

```python
euclidean_distances([query_embedding], document_embedding)
```

```python
import faiss
from langchain_community.vectorstores import FAISS
from langchain_community.docstore.in_memory import InMemoryDocstore
```

```python
index=faiss.IndexFlatL2(384) #size of embedding as per the model used, while creating the index
```

```python
index
```

```python
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_texts(["AI is future", "AI is powerful", "Dogs are cute"])
```

The above o/p are UUIDs. Here we've stored the docs inside the vector stores.

```python
vector_store.index_to_docstore_id
```

```python
results=vector_store.similarity_search("Tell me about AI", k=2)
results
```

```python
from langchain_core.documents import Document

document_1 = Document(
    page_content="I had chocolate chip pancakes and scrambled eggs for breakfast this morning.",
    metadata={"source": "tweet"},
)

document_2 = Document(
    page_content="The weather forecast for tomorrow is cloudy and overcast, with a high of 62 degrees.",
    metadata={"source": "news"},
)

document_3 = Document(
    page_content="Building an exciting new project with langchain - come check it out!.",
    metadata={"source": "tweet"},
)

document_4 = Document(
    page_content="Robbers broke into the city bank and stole $1 million in cash.",
    metadata={"source": "news"},
)

document_5 = Document(
    page_content="Wow! that was an amazing movie. I can't wait to see it again.",
    metadata={"source": "tweet"},
)

document_6 = Document(
    page_content="Is the new iPhone worth the price? Read this review to find out.",
    metadata={"source": "website"},
)

document_7 = Document(
    page_content="The top 10 soccer players in the world right now.",
    metadata={"source": "website"},
)

document_8 = Document(
    page_content="LangGraph is the best framework for building stateful, agentic applications!.",
    metadata={"source": "tweet"},
)

document_9 = Document(
    page_content="The stock market is down 500 points today due to the fears of recession.",
    metadata={"source": "news"},
)

document_10 = Document(
    page_content="I have a bad feeling I am going to get deleted :(",
    metadata={"source": "tweet"},
)

documents=[
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
    document_6,
    document_7,
    document_8,
    document_9,
    document_10,
]
```

```python
index=faiss.IndexFlatIP(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
vector_store.add_documents(documents=documents)
```

```python
result=vector_store.similarity_search(
    "Langchain provides abstractions to make working with LLMs easy.",
    # k=2,
    # filter={"source":{"$eq": "tweet"}}
    filter={"source":"news"}
)
```

```python
result
```

```python
result[0].metadata
```

```python
result[0].page_content
```

```python
## Using the vector store as retriever in RAG pipeline

retriever=vector_store.as_retriever(search_kwargs={"k":3})
```

```python
retriever.invoke("Langchain provides abstractions to make working with LLMs easy.")
```

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
FILE_PATH="llama2-bf0a30209b224e26e31087559688ce81.pdf"
```

```python
loader=PyPDFLoader(FILE_PATH)
```

```python
loader.load()
```

```python
len(loader.load())
```

```python
pages=loader.load()
```

```python
# Chunking
from langchain_text_splitters import  RecursiveCharacterTextSplitter
splitter=RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)
```

```python
split_docs=splitter.split_documents(pages)
len(split_docs)
```

```python
# split_docs
```

```python
index=faiss.IndexFlatIP(384)
vector_store=FAISS(
    embedding_function=embeddings,
    index=index,
    docstore=InMemoryDocstore(),
    index_to_docstore_id={},
)
```

```python
vector_store.add_documents(documents=split_docs)
```

```python
retriever=vector_store.as_retriever(
    search_kwargs={"k": 10}
)
```

```python
retriever.invoke("What is a llama model?")
```
