#### Embedding Techniques

to convert text to vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ["OPENAI_API_KEY"]=os.getenv('OPENAI_API_KEY')
```

```python
from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
```

```python
embeddings
```

```python
text="This is a tutorial on OPENAI embeddings"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

The openAI generates embeddings to a fixed length of 3072. We can also control the number of dimensions also.

```python
from langchain_openai import OpenAIEmbeddings

embeddings = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
text="This is a tutorial on OPENAI embeddings"
query_result=embeddings.embed_query(text)
query_result
```

```python
len(query_result)
```

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("speech.txt")
docs=loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=200)
final_documents = text_splitter.split_documents(docs)
len(final_documents)
```

```python
final_documents[0]
```

```python
embeddings.embed_query(final_documents[0].page_content)
```
