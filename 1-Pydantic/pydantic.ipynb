- dataclass in python is used as a decorator, used on other class to define some attributes

```python
from dataclasses import dataclass
```

```python
@dataclass
class Person:
    name:str
    age:int
    city:str

```

```python
person=Person(name="Swastika", age=30, city="Ahmedabad")
print(person)
```

```python
from pydantic import  BaseModel

class Person1(BaseModel):
    name:str
    age:int
    city:str
```

- for data validation we need to inherit BaseModel

```python
person = Person1(name="Swastika", age=28, city="Ahmedabad")
print(person)
```

```python
person = Person1(name="Swastika", age=28, city=45)
print(person)
```

```python
person = Person1(name="Swastika", age=28, city='45')
print(person)
```

### Model with Optional Value

```python
from typing import Optional
```

```python
class Employee(BaseModel):
    id:int
    name:str
    department:str
    salary:Optional[float]=None
    is_active:Optional[bool]=True

emp1=Employee(id=1, name="Swastika", department="AI")
print(emp1)
```

Wherever typecasting is possible, it will be automatically taken care of. Eg. converting integer str to str if var defined as str

```python
emp2=Employee(id=1, name="Swastika", department="AI", salary='10000')
print(emp2)
```

Definition:
    - Indicates the field can be None
    - Default value (= None or = True): Makes the field optional
    - Required fields must still be provided
    - Pydantic validates types even for optional fields when values are provided

```python
from typing import List
class Classroom(BaseModel):
    room_number:int
    students:List[str]
    capacity:int
    
```

```python
classroom=Classroom(room_number=1, students=["Swastika", "Raj"], capacity=10)
print(classroom)
```

#### Models with Nested Models

Create complex structures with nested models

- One class inheriting the Base Class. The same class can be used in another class.

```python
from pydantic import BaseModel

class Address(BaseModel):
    street:str
    city:str
    zip_code:str

class Customer(BaseModel):
    customer_id:int
    name:str
    address:Address
```

```python
customer=Customer(customer_id=1, name="Swastika", 
                  address=Address(street="123 Main St", city="Boston", zip_code="12345"))
print(customer)
```

```python
customer=Customer(customer_id=1, name="Swastika", 
                  address=Address(street="123 Main St", city=123, zip_code="12345"))
print(customer)
```

#### Pydantic Fields: Customization and Constraints

The Field function in Pydantic enhances model fields beyond basic type hints by allowing you to specify validation rules, default values, aliases, and more. Here's a comprehensive tutorial with examples.

```python
from pydantic import BaseModel, Field

class Item(BaseModel):
    name:str=Field(min_length=2, max_length=50)
    price:float=Field(gt=0, lt=10000)
    quantity:int=Field(ge=0)
    
```

```python
item=Item(name="Laptop", price=1000, quantity=1)
print(item)
```

```python
item=Item(name="Laptop", price=100000, quantity=1)
print(item)
```

```python
class User(BaseModel):
    id:int
    username:str=Field(description="Unique username")
    age:int=Field(default=18, description="Age default to 18.")
    email:str=Field(default_factory=lambda: "user@example.com", description="Default email")
    
user=User(id=1, username="Swastika")
print(user)
```

```python
user=User(id=1, username="Swastika", age=28, email="swastika@gmail.com")
print(user)
```
