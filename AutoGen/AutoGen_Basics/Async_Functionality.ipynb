#### Async Functionality in Python

```python
## Synchronous or Normal way of Coding
def fun1():
    print('Hello! I am first function')

def fun2():
    print('Hello! I am second function')

fun1()
fun2()

print("Done")
```

```python
import time

def brew_coffee():
    print("Brewing coffee...")
    time.sleep(3)
    print("Coffee Ready")

def toast_bagel():
    print("Toasting bagel...")
    time.sleep(2)
    print("Bagel Ready")

def main():
    start=time.time()
    coffee = brew_coffee()
    bagel=toast_bagel()

    end=time.time()
    print(f"Total time taken: {end-start: .2f} seconds")

main()
    
    
```

#### Asynchronous Version

```python
import time
import asyncio

async def brew_coffee():
    print("Brewing coffee...")
    await asyncio.sleep(3)
    print("Coffee Ready")

async def toast_bagel():
    print("Toasting bagel...")
    await asyncio.sleep(2)
    print("Bagel Ready")

async def main():
    start=time.time()
    coffee = brew_coffee()
    bagel=toast_bagel()

    results = await asyncio.gather(coffee, bagel)

    end=time.time()
    print(f"Total time taken: {end-start: .2f} seconds")

await main()
    
    
```
