```python
# print(api_key)
```

```python
## Connecting to the model
## Created brain
model_client = OpenAIChatCompletionClient(
    model="gpt-4o-mini",
    api_key=api_key
)
```

```python
## Created the digital human and provided its brain (model_client) 
agent_1 = AssistantAgent(
    name="my_assistant",
    model_client=model_client,
)
```

```python
result = await agent_1.run(task="tell me something about you?")
```

### Using different models with Autogen

#### Ollama-local hosted

```python
from autogen_ext.models.ollama import OllamaChatCompletionClient
from autogen_core.models import UserMessage

# Initialize the client
ollama_client = OllamaChatCompletionClient(model="llama3")
response = await ollama_client.create([UserMessage(content="tell me something about you?", source='user')])
print(response)
await ollama_client.close()
```

```python
model_client = OpenAIChatCompletionClient(
    model="gemini-2.0-flash",
    api_key=""
)

response = await model_client.create([UserMessage(content="What is the capital of France?", source='user')])
print(response)
await model_client.close()
```

#### OpenRouter

```python
open_router_api_key = os.getenv("OPEN_ROUTER_API_KEY")
```

```python
# print(open_router_api_key)
```

```python

open_router_model_client = OpenAIChatCompletionClient(
    base_url="https://openrouter.ai/api/v1",
    api_key=open_router_api_key,
    model="gemma-4-31b-it",
    model_info={
        "family": "gemma",
        "function_calling": True,
        "json_output": False,
        "vision": False
    }
)

response = await open_router_model_client.create([UserMessage(content="What is the capital of France?", source='user')])
print(response)
```

```python
response_2 = await open_router_model_client.create([UserMessage(content="What is the capital of India?", source='user')])
print(response_2)
```
